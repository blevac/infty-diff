```python
import numpy as np
import torch
import glob
from mri_dataloader import fastMRI_brain
from torch.utils.data import Dataset, DataLoader
```

```python
file_list = glob.glob('/csiNAS2/slow/brett/fastmri_brain_preprocessed_256/AXT2/**/*.npy')
print(len(file_list))
```

```python
# create a dataset object with fasMRI_brain class
dataset = fastMRI_brain(file_list)
# create a dataloader object
dataloader = DataLoader(dataset, batch_size=1, shuffle=True)
# iterate through the dataloader
for i, data in enumerate(dataloader):
    print(data.shape)
    if i == 0:
        break
```

```python
# plot image from dataloader
import matplotlib.pyplot as plt
cplx_img = data[0,0,:,:] + 1j*data[0,1,:,:]
plt.figure()
plt.imshow(abs(cplx_img),cmap='gray')
```
