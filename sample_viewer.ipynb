```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
img_sz = 131
file = '/home/blevac/infty-diff/checkpoints/brain_256_sampler_samples_%d_100/00000.pt'%(img_sz)
cont = torch.load(file)
print(cont.shape)
```

```python
def complex_conv(img):
    img_out = img[0] + 1j*img[1]
    return img_out
```

```python
cplx_img = complex_conv(cont)
plt.figure()
plt.imshow(abs(cplx_img),cmap='gray')
```
